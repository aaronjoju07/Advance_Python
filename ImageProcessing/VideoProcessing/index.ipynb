```python
import numpy as np
import cv2 as cv
```

```python
cap=cv.VideoCapture('/Users/aaronjoju/Documents/Advance_Python/ImageProcessing/VideoProcessing/m.mp4')
```

```python
while(cap.isOpened()):
    ret,frame = cap.read()
    print(ret,frame)
    if ret==True:
        cv.imshow('Frame',frame)
        if cv.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv.destroyAllWindows()
```

```python
import os
cap=cv.VideoCapture('/Users/aaronjoju/Downloads/myvid.mov')

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print('ERROR:')

curr_frame = 0

while(True):
    ret,frame = cap.read()
    if ret:
        name = './data/frame' + str(curr_frame)+'.jpg'
        cv.imwrite(name,frame)
        curr_frame +=1
    else:
        break
cap.release()
cv.destroyAllWindows()
```
