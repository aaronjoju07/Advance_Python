# Identify the suitable variables from the given dataset and create the following 2D graphs using Python libraries like Matplotlib etc.


```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Advance Python/Dataset/UScereal.csv')
data.head()
```

## Bar Charts using 2 variables

```python
fig = px.bar(data, x='mfr', y='protein')
fig.show()
```

## Bar Charts using 3 variables

```python
fig = px.bar(data, x="mfr", y="protein", color="vitamins", title="Long-Form Input")
fig.show()
```

## Scatter Plots using 2 variables

```python
fig = px.scatter(x=data.sodium)
fig.show()
```

## Scatter Plots using 3 variables

```python
fig = px.scatter_3d(data, x=data.mfr, y=data.protein, z=data.calories)

# Updating figure layout
fig.update_layout(scene=dict(
                    xaxis_title='mfr',
                    yaxis_title='protein',
                    zaxis_title='calories')
)
fig.show()
```

## Histogram using 2 variables

```python
fig = px.histogram(data, x="vitamins")
fig.show()
```

## Pie chart

```python
manufacturer_counts = data['mfr'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Manufacturer Distribution')
plt.show()
```

## Pair plot

```python

df = pd.DataFrame(data)

# Filter out non-numerical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

sns.pairplot(numeric_cols)
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```

## Bubble plot

```python
sns.scatterplot(data=df, x='calories', y='protein', size='sodium', sizes=(20, 200), hue='mfr', palette='Set1', alpha=0.7)
plt.title('Bubble Plot of Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['calories','sodium', 'sugars', 'potassium']])
plt.title('Box Plot of Nutritional Values')
plt.xlabel('Nutritional Values')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

## Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[['calories','sodium','sugars', 'potassium']])
plt.title('Violin Plot of Nutritional Values')
plt.xlabel('Nutritional Values')
plt.ylabel('Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

## Area Plot

```python
fig = px.area(df, x="mfr", y="sodium", color="mfr")
fig.show()
```

## Line Plot

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Name'], df['calories'], marker='o', label='Calories')
plt.plot(df['Name'], df['protein'], marker='o', label='Protein')
plt.plot(df['Name'], df['fat'], marker='o', label='Fat')
plt.plot(df['Name'], df['sodium'], marker='o', label='Sodium')
plt.title('Line Plot of Nutritional Values')
plt.xlabel('Product Name')
plt.ylabel('Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()
plt.grid(True)  # Add grid for better visualization
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```
