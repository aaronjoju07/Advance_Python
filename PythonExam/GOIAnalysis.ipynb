```python
import pandas as pd

data = pd.read_csv('GOI.csv')
data.info()
```

```python
data.head()
```

```python
# Explore the literacy rates across the union territories based on year 2001 and 2011.
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 16))
ax1 = fig.add_subplot(221, projection='3d')

A_N_ISLANDS = data[data['Country/ States/ Union Territories Name'] == 'A & N ISLANDS']
CHANDIGARH = data[data['Country/ States/ Union Territories Name'] == 'CHANDIGARH']
D_N_HAVELI = data[data['Country/ States/ Union Territories Name'] == 'D & N HAVELI']
DAMAN_DIU = data[data['Country/ States/ Union Territories Name'] == 'DAMAN & DIU']
LAKSHADWEEP = data[data['Country/ States/ Union Territories Name'] == 'LAKSHADWEEP']
NCT_OF_DELHI = data[data['Country/ States/ Union Territories Name'] == 'NCT OF DELHI']
PUDUCHERRY = data[data['Country/ States/ Union Territories Name'] == 'PUDUCHERRY']

ax1.bar(A_N_ISLANDS['Literacy Rate (Persons) - Total - 2001'], A_N_ISLANDS['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=0, color='red')
ax1.bar(CHANDIGARH['Literacy Rate (Persons) - Total - 2001'], CHANDIGARH['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=1, color='blue', label='CHANDIGARH')
ax1.bar(D_N_HAVELI['Literacy Rate (Persons) - Total - 2001'], D_N_HAVELI['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=2, color='green', label='D & N HAVELI')
ax1.bar(DAMAN_DIU['Literacy Rate (Persons) - Total - 2001'], DAMAN_DIU['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=3, color='purple', label='DAMAN & DIU')
ax1.bar(LAKSHADWEEP['Literacy Rate (Persons) - Total - 2001'], LAKSHADWEEP['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=4, color='orange', label='LAKSHADWEEP')
ax1.bar(NCT_OF_DELHI['Literacy Rate (Persons) - Total - 2001'], NCT_OF_DELHI['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=5, color='pink', label='NCT OF DELHI')
ax1.bar(PUDUCHERRY['Literacy Rate (Persons) - Total - 2001'], PUDUCHERRY['Literacy Rate (Persons) - Total - 2011'], zdir='y', zs=6, color='brown', label='PUDUCHERRY')


ax1.set_xlabel('Literacy Rate (Persons) - Total - 2001')
ax1.set_ylabel('Literacy Rate (Persons) - Total - 2011')
ax1.set_zlabel('Union Territories')
ax1.set_title('Literacy Rates Across Union Territories (2001 vs. 2011)')
ax1.legend()

plt.show()
```

```python
states_data = data[data['Category'] == 'State']
rural_literacy_2001 = states_data['Literacy Rate (Persons) - Rural - 2001']
urban_literacy_2001 = states_data['Literacy Rate (Persons) - Urban - 2001']
rural_literacy_2011 = states_data['Literacy Rate (Persons) - Rural - 2011']
urban_literacy_2011 = states_data['Literacy Rate (Persons) - Urban - 2011']

fig = plt.figure(figsize=(25, 20))
ax = fig.add_subplot(221, projection='3d')

bar_width = 0.4

ax.bar(states_data.index - bar_width/2, rural_literacy_2001, zs=0, zdir='y', width=bar_width, color='blue', label='Rural 2001')
ax.bar(states_data.index + bar_width/2, urban_literacy_2001, zs=0, zdir='y', width=bar_width, color='cyan', label='Urban 2001')

ax.bar(states_data.index - bar_width/2, rural_literacy_2011, zs=1, zdir='y', width=bar_width, color='red', label='Rural 2011')
ax.bar(states_data.index + bar_width/2, urban_literacy_2011, zs=1, zdir='y', width=bar_width, color='yellow', label='Urban 2011')

ax.set_xlabel('States')
ax.set_ylabel('Literacy Rate (%)')
ax.set_zlabel('Number of States')
ax.set_title('3D Bar Chart of Urban vs. Rural Literacy Rates (2001 and 2011)')
ax.set_xticks(states_data.index)
ax.set_xticklabels(states_data['Country/ States/ Union Territories Name'], rotation=45, ha='right')
ax.legend()

plt.show()
```

```python
# Create a Indian state map to highlight level variations of Total Literacy of the year
# 2011. Levels can be measured using following:(5 Marks)
# Poor Min-25% , Fair 26% to 50%,
# Good 51% to 75%, Excellent  76% to Max
```

```python
bins = [0, 25, 50, 75, 100]
labels = ['Poor', 'Fair', 'Good', 'Excellent']

merged_data['Literacy Levels'] = pd.cut(merged_data['Literacy Rate (Persons) - Total - 2011'], bins=bins, labels=labels, include_lowest=True)
fig, ax = plt.subplots(figsize=(12, 8))
merged_data.plot(column='Literacy Levels', cmap='YlGnBu', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

legend_labels = {label: f'{bins[i]}% - {bins[i+1]}%' for i, label in enumerate(labels)}
ax.legend(title='Total Literacy Levels (2011)', labels=legend_labels.values(), loc='upper left')

plt.show()
```
