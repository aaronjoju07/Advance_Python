```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

df = pd.read_csv('GOI.csv')
df['Category'].unique()
```

```python
ut_data = df[df['Category'] == 'Union Territory']
ut_data
```

```python
fig = plt.figure(figsize=(20, 20))

ax1 = fig.add_subplot(211, projection='3d')
ax1.bar(ut_data['Country/ States/ Union Territories Name'], ut_data['Literacy Rate (Persons) - Total - 2001'], zs=2001, zdir='y', color='y', alpha=0.7, label='Total Literacy Rate - 2001')
ax1.bar(ut_data['Country/ States/ Union Territories Name'], ut_data['Literacy Rate (Persons) - Total - 2011'], zs=2011, zdir='y', color='g', alpha=0.7, label='Total Literacy Rate - 2011')
ax1.set_xlabel('Union Territories')
ax1.set_ylabel('Year')
ax1.set_zlabel('Literacy Rate (%)')
ax1.set_title('Total Literacy Rates Across Union Territories (2001 and 2011)')
ax1.legend()

plt.show()
```

```python
states_data = df[df['Category'] == 'State']
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(states_data['Literacy Rate (Persons) - Rural - 2011'], 
           states_data['Literacy Rate (Persons) - Urban - 2011'], 
           zs=0,  
           s=50,  
           c='g', 
           label='2011')

ax.scatter(states_data['Literacy Rate (Persons) - Rural - 2001'], 
           states_data['Literacy Rate (Persons) - Urban - 2001'], 
           zs=0, 
           s=50,
           c='y', 
           alpha=0.7, 
           label='2001')

# Add labels and title
ax.set_title('Relationship between Urban and Rural Literacy Rates (2001 and 2011)')
ax.set_xlabel('Rural Literacy Rate (%)')
ax.set_ylabel('Urban Literacy Rate (%)')
ax.set_zlabel('Year (0 for 2011, 1 for 2001)')
ax.legend()
ax.grid(True)

# Show the plot
plt.show()
```

```python
merged_data = india_gdf.merge(df[df['Category'] == 'State'], how='left', left_on='STATE', right_on='Country/ States/ Union Territories Name')

merged_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

merged_data.plot(column='Literacy Rate (Persons) - Total - 2011', cmap='YlGnBu', ax=ax, legend=True)
ax.set_title('Total Literacy Rate India')

plt.show()
```

```python
bins = [0, 25, 50, 75, 100]
labels = ['Poor', 'Fair', 'Good', 'Excellent']

merged_data['Literacy Levels'] = pd.cut(merged_data['Literacy Rate (Persons) - Total - 2011'], bins=bins, labels=labels, include_lowest=True)
fig, ax = plt.subplots(figsize=(12, 8))
merged_data.plot(column='Literacy Levels', cmap='YlGnBu', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

plt.show()
```
