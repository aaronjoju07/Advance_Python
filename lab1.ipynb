```python
import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Advance Python/Dataset/Q8_MortalityDataset.csv')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
alive = data[data['MORT'] == 'alive']
dead = data[data['MORT'] == 'dead']
ax.scatter(alive['AGE'],alive['HEIGHT'],alive['WEIGHT'],c='green',marker='x',label='Alive')
ax.scatter(dead['AGE'],dead['HEIGHT'],dead['WEIGHT'],c='red',marker='o',label='Dead')
ax.set_title('MORT ')
ax.set_xlabel('AGE')
ax.set_ylabel('HEIGHT')
ax.set_zlabel('WEIGHT')
ax.legend()
plt.show()
```

```python
# Two 3D scatter plots
fig = plt.figure(figsize=(20, 18))

# Plot for alive individuals
ax3 = fig.add_subplot(221)
ax3.scatter(alive['HEIGHT'], alive['WEIGHT'],c='red', label='Alive')
ax3.set_xlabel('Height')
ax3.set_ylabel('Weight')

ax3.set_title('2D Scatter Plot for Alive Individuals')
ax3.legend()

# Plot for dead individuals
ax4 = fig.add_subplot(222)
ax4.scatter(dead['HEIGHT'], dead['WEIGHT'], c='green' ,label='Dead')
ax4.set_xlabel('Height')
ax4.set_ylabel('Weight')
ax4.set_title('2D Scatter Plot for Dead Individuals')
ax4.legend()

# Plot for alive individuals
ax1 = fig.add_subplot(223, projection='3d')
ax1.scatter(alive['HEIGHT'], alive['WEIGHT'], zs=0,c='red', label='Alive')
ax1.set_xlabel('Height')
ax1.set_ylabel('Weight')
ax1.set_zlabel('Mortality Status')
ax1.set_title('3D Scatter Plot for Alive Individuals')
ax1.legend()

# Plot for dead individuals
ax2 = fig.add_subplot(224, projection='3d')
ax2.scatter(dead['HEIGHT'], dead['WEIGHT'], zs=1,c='green' ,label='Dead')
ax2.set_xlabel('Height')
ax2.set_ylabel('Weight')
ax2.set_zlabel('Mortality Status')
ax2.set_title('3D Scatter Plot for Dead Individuals')
ax2.legend()

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
a = data[data['BLOOD'] == 'a']
b = data[data['BLOOD'] == 'b']
o = data[data['BLOOD'] == 'o']
ab = data[data['BLOOD'] == 'ab']
ax.bar(a['AGE'],a['CHOL'],zdir='y',zs=0,color='pink',label='a')
ax.bar(b['AGE'],b['CHOL'],zdir='y',zs=1,color='red',label='b')
ax.bar(ab['AGE'],ab['CHOL'],zdir='y',zs=2,color='orange',label='ab')
ax.bar(o['AGE'],o['CHOL'],zdir='y',zs=3,color='purple',label='b')
ax.set_title('3D Bar plot')
ax.set_xlabel('AGE')
ax.set_ylabel('Blood')
ax.set_zlabel('CHOL')
ax.legend()
plt.show()
```
