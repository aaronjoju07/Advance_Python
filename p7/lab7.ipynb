### Text Preprocessing

#### Perform tokenization on the text data to split it into individual words or tokens using four different methods and use better results for further analysis.

```python
import nltk
# nltk.download('all')
from nltk.tokenize import sent_tokenize,word_tokenize,WordPunctTokenizer,TreebankWordTokenizer,RegexpTokenizer
```

##### word_tokenize

```python
word_tokenize_out = word_tokenize(data)
print(len(word_tokenize_out))
word_tokenize_out
```

##### WordPunctTokenizer

```python
WordPunctTokenizer_out = WordPunctTokenizer().tokenize(data)
print(len(WordPunctTokenizer_out))
WordPunctTokenizer_out
```

##### TreebankWordTokenizer

```python
TreebankWordTokenizer_out = TreebankWordTokenizer().tokenize(data)
print(len(TreebankWordTokenizer_out))
TreebankWordTokenizer_out
```

```python
tokenizee = RegexpTokenizer("[\w']+")
RegexpTokenizer_out =tokenizee.tokenize(data)
print (len(RegexpTokenizer_out)) 
RegexpTokenizer_out
```

#### Remove any stopwords (common words with little semantic meaning) from the text.

```python
from nltk.corpus import stopwords
```

```python
# Load English stopwords
stop_words = set(stopwords.words('english'))
```

```python
# Remove stopwords
filtered_tokens = [word for word in WordPunctTokenizer_out if word.lower() not in stop_words]
# filtered_text = ' '.join(filtered_tokens)

# print(filtered_tokens)
```

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word.lower()) for word in filtered_tokens]
print(len(lemmatized_tokens))  # Output
lemmatized_tokens
```

#### Clean the text by removing special characters, punctuation, and numbers.

```python
import re
cleaned_tokens = [re.sub(r'[^a-zA-Z]', '', token) for token in lemmatized_tokens if token.isalpha()]
print(len(cleaned_tokens))
cleaned_tokens
```

### Text Visualization

```python
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
import pandas as pd
```

#### Plot the distribution of word frequencies in the text data(Line Chart).

```python
fd = FreqDist(cleaned_tokens)

# Plot the frequency distribution
plt.figure(figsize=(10, 6))
fd.plot(10, title='Frequency Distribution of Lemmatized Tokens')
plt.show()
```

#### Visualize the most frequent 10 words or terms in the text data using bar charts.

```python

y =fd.most_common(10)
print(y)
df = pd.DataFrame(y,columns=['Word', 'Frequency'])
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Word'], df['Frequency'])
plt.title('Top 10 Word Frequencies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
