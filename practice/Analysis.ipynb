```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('UScereal.csv')
data.head()
```

```python
data.info()
```

```python

x=data['fat']
y=data['protein']
plt.plot(x,y,)
plt.title('Line Plot Fat v/s proteins')
plt.xlabel('fat')
plt.ylabel('proteins')
plt.show()
```

```python
plt.scatter(x, y,color='black',marker='x')
plt.title('Scatter Plot Fat v/s proteins')
plt.xlabel('X-Fat')
plt.ylabel('Y-proteins')
```

```python
values = [2,3,10,12]
plt.bar(x,y)
plt.show()
```

```python
plt.hist(x,bins=30,color='purple')
plt.show()
```

```python
vitamins = data['vitamins'].unique()
vitamins_count = data['vitamins'].value_counts()
plt.pie(vitamins_count,startangle=90,colors=['purple','green','blue'])
plt.title('Pie Chart  of Vitamin Distribution')
plt.legend(vitamins)
plt.show()
```

```python
iris = pd.read_csv('Iris.csv')
iris.info()
```

```python
from mpl_toolkits import mplot3d
```

```python
iris['Species'].unique()
```

```python
# Scatter
fig = plt.figure(figsize= (10,20))

ax = fig.add_subplot(221,projection= '3d')
Iris_setosa = iris[iris['Species'] == 'Iris-setosa ']
Iris_virginica = iris[iris['Species'] == 'Iris-virginica']
Iris_versicolor = iris[iris['Species'] == 'Iris-versicolor']
plt.scatter(Iris_setosa['SepalLengthCm'],Iris_setosa['SepalWidthCm'],Iris_setosa['PetalLengthCm'],marker='o',label='Iris_setosa')
plt.scatter(Iris_versicolor['SepalLengthCm'],Iris_versicolor['SepalWidthCm'],Iris_versicolor['PetalLengthCm'],marker='x',label='Iris_versicolor')
plt.scatter(Iris_virginica['SepalLengthCm'],Iris_virginica['SepalWidthCm'],Iris_virginica['PetalLengthCm'],marker='.',label='Iris_virginica')
ax.legend()

# Bar
ax = plt.axes(222,projection= '3d')
plt.bar(Iris_setosa['SepalLengthCm'],Iris_setosa['SepalWidthCm'],zdir='y',zs=0,label='Iris_setosa')
plt.bar(Iris_versicolor['SepalLengthCm'],Iris_versicolor['SepalWidthCm'],zdir='y',zs=1,label='Iris_versicolor')
plt.bar(Iris_virginica['SepalLengthCm'],Iris_virginica['SepalWidthCm'],zdir='y',zs=2,label='Iris_virginica')
ax.legend()
plt.show()
```

### BaseMap

```python
from mpl_toolkits.basemap import Basemap
```

```python
figure = plt.figure(figsize=(10,4))
map = Basemap(projection='ortho',resolution='c',lat_0=20.5937,lon_0=78.9629)
map.bluemarble(scale=0.5)
```

```python
figure = plt.figure(figsize=(5,5))
map = Basemap(projection='lcc',resolution='l',width=1200000,height=900000,lat_0=20.5937,lon_0=78.9629)
map.drawcoastlines()
map.drawcounties()
map.drawrivers()
map.drawstates()
map.fillcontinents(color='lightgray',lake_color='aqua')
map.drawmapboundary(fill_color='aqua')

x,y=map(80,22)
plt.plot(x,y,'ok',markersize=5)
plt.text(x,y,'India',fontsize=15)
```

```python
figure = plt.figure(figsize=(12,12))
map = Basemap()
map.drawcoastlines()
map.fillcontinents(color='orange',lake_color='aqua')
map.drawmapboundary(fill_color='aqua')
map.drawcounties(color='brown',linewidth=1.0)
map.drawrivers(color='blue',linewidth=0.5)
map.drawmeridians(range(0,360,40),color='k',labels=[0,0,1,1])
map.drawparallels(range(-90,150,40),color='k',labels=[1,0,0,0])
plt.show()
```
