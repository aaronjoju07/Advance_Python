```python
import matplotlib
import pandas as pd
import contextily as ctx
import geopandas as gpd

```

```python


world_gfd = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world_gfd

```

```python


world_gfd.crs

```

```python


world_gfd['pop_density']= world_gfd.pop_est / world_gfd.area * 10**6
world_gfd.sort_values(by='pop_density',ascending = False)


```

```python


figsize =(20, 11)
world_gfd.plot('pop_density', legend=True, figsize=figsize)
```

```python
norm=matplotlib.colors.LogNorm(vmin=world_gfd.pop_density.min(),vmax=world_gfd.pop_density.max())
world_gfd.to_crs("epsg:4326").plot(column="pop_density",figsize=figsize, legend=True, norm=norm)
```
